##**KPI Analysis**


```python
# Total Sales

Total_Sales=df['Sales'].sum()
print(Total_Sales)
```

```python
# Average Sales

Avg_Sales=df['Sales'].mean()
print(Avg_Sales)
```

```python
# Number of Item sold

No_Item_Sold=df['Item Identifier'].count()
print(No_Item_Sold)
```

```python
# Average Rating

Avg_Rating=df['Rating'].mean()
print(Avg_Rating)
```

### **Displaying KPI's**

```python
# Display All KPI's

print(f"Total Sales :$ {Total_Sales:,.0f}")
print(f"Average Sales :$ {Avg_Sales:,.1f}")
print(f"Items Sold: {No_Item_Sold:,.0f}")
print(f"Average Rating:{Avg_Rating:,.1f}")
```

##**Charts Requirements**

####**Total sales by Fat Content**

```python
sales_by_fat=df.groupby('Item Fat Content')['Sales'].sum()

#plot graph

plt.pie(sales_by_fat, labels= sales_by_fat.index , autopct='%.1f%%',startangle=90)
plt.title('Total Sales by Fat Content')
plt.show()
```

#### From this pie chart we come to know that items with Low Fat has 64.6% total Sales and 35.4% Total Sales for  regular

####**Total Sales by Item Type**

```python
Sales_by_item_type= df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)


# plotting graph

plt.figure(figsize=(9,7))

bars=plt.bar(Sales_by_item_type.index,Sales_by_item_type.values,color='green')

plt.title('Total Sales by Item Type')
plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')

for bar in bars:
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),
           f'{bar.get_height():,.0f}',ha='center',va='bottom',fontsize=8)

plt.tight_layout()
plt.show()
```

####**Fat Content by Outlet for Total Sales**

```python
grouped= df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
grouped= grouped[['Low Fat','Regular']]

ax=grouped.plot(kind='bar',figsize=(10,5),title='Outlet Tier by Fat Content')
plt.ylabel('Total Sales')
plt.xlabel('Outlet Location Tier')

for bar in ax.patches:
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),
           f"{bar.get_height():,.0f}",ha='center',va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

####**Total Sales by Outlet Establishment**

```python
sales_by_year=df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(8,5))

plt.plot(sales_by_year.index,sales_by_year.values,marker='o')
plt.title('Total Sales by Outlet Establishment ')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')

for x,y in zip(sales_by_year.index,sales_by_year.values):
  plt.text(x,y,f"{y:,.0f}",ha='center',va='bottom',fontsize=8)

plt.tight_layout()
plt.show()
```

####**Total Sales by Outlet Size**

```python
Sales_by_size=df.groupby('Outlet Size')['Sales'].sum()

plt.figure(figsize=(5,5))
plt.pie(Sales_by_size,labels=Sales_by_size.index,autopct='%.f%%')

plt.title('Sales By outlet Size')
plt.show()
```
